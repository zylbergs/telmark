```python
# Step 1: Install necessary libraries
# You can install these libraries using pip if you haven't already:
# pip install psycopg2 kafka-python

import psycopg2
from kafka import KafkaProducer
import json

# Step 2: Set up the PostgreSQL connection
db_username = 'your_username'
db_password = 'your_password'
db_host = 'your_host'
db_port = 'your_port'
db_name = 'your_database_name'
table_name = 'your_table_name'

conn = psycopg2.connect(
    dbname=db_name,
    user=db_username,
    password=db_password,
    host=db_host,
    port=db_port
)

# Step 3: Read data from the PostgreSQL table
cur = conn.cursor()
cur.execute(f"SELECT * FROM {table_name}")
rows = cur.fetchall()

# Step 4: Set up the Kafka producer
kafka_bootstrap_servers = 'your_kafka_bootstrap_servers'
kafka_topic = 'your_kafka_topic'

producer = KafkaProducer(
    bootstrap_servers=kafka_bootstrap_servers,
    value_serializer=lambda v: json.dumps(v).encode('utf-8')
)

# Step 5: Send the data to the Kafka topic
for row in rows:
    producer.send(kafka_topic, row)

# Close the PostgreSQL connection
cur.close()
conn.close()

# Close the Kafka producer
producer.close()

print(f"Data from table '{table_name}' has been sent to Kafka topic '{kafka_topic}'.")
```
